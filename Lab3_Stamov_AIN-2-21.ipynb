```python
import numpy as np
import matplotlib.pyplot as plt
```

<h4>2.	Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры</h4>

```python

a = 1.0
b = -2.0
c = 1.0


x = np.arange(-10, 10, 0.1)


y = a * x**2 + b * x + c


plt.figure(figsize=(8, 6)) 

plt.plot(x, y, label=f'y = {a}x^2 + {b}x + {c}', color='b') 
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('График квадратичной функции')
plt.grid(True)
plt.legend()

plt.show()
```

<h4>3.	Построить график произвольной гармонической функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры</h4>

```python
A = 2.0
B = 1.5
C = 0.5

x2 = np.linspace(0, 10, 400)

y2 = A * np.sin(B * x2 + C)

plt.figure(figsize=(8, 6))

plt.plot(x2, y2, label=f'y = {A} * sin({B} * x + {C})', color='b')
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('График гармонической функции')
plt.grid(True)
plt.legend()

plt.show()
```

<h4>4.	Вывести графики п.2 и 3 в разные поля при помощи функции subplot()</h4>

```python
plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
plt.plot(x, y, label=f'y = {a}x^2 + {b}x + {c}', color='b')
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('График квадратичной функции')
plt.grid(True)
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(x2, y2, label=f'y = {A} * sin({B} * x + {C})', color='r')
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('График гармонической функции')
plt.grid(True)
plt.legend()


plt.tight_layout()


plt.show()
```

<h4>5.	Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.</h4>

```python
plt.figure(figsize=(10, 6))

plt.plot(x, y, label=f'y = {a}x^2 + {b}x + {c}', color='b')
plt.xlabel('X-ось')
plt.ylabel('Y-ось (масштаб 0 до 120)')
plt.grid(True)
plt.legend()

ax2 = plt.gca().twinx()
ax2.set_ylim(-2, 2)
ax2.set_ylabel('Y-ось (масштаб -2 до 2)')

ax2.plot(x2, y2, label=f'y = {A} * sin({B} * x + {C})', color='r')
ax2.legend()

plt.tight_layout()
plt.show()

```

<h4>6.	Построить график функции, заданной параметрически, </h4>

```python
t = np.linspace(0, 4 * np.pi, 1000)


x = np.cos(3 * t)
y = np.sin(2 * t)


plt.figure(figsize=(8, 6))

plt.plot(x, y, label='Параметрическая функция: $x(t) = \cos(3t), y(t) = \sin(2t)$', color='b')
plt.xlabel('X(t)')
plt.ylabel('Y(t)')
plt.title('График параметрически заданной функции')
plt.grid(True)
plt.legend()


plt.show()
```

<h4>7.	Построить столбиковую диаграмму по данным статистического комитета КР по прожиточному минимуму, представленному в таблице</h4>

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
min_expenses = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]

plt.figure(figsize=(10, 6))
plt.bar(years, min_expenses, color='skyblue')
plt.xlabel('Год')
plt.ylabel('Прожиточный минимум')
plt.title('Прожиточный минимум в КР (2012-2022)')
plt.xticks(years, rotation=45)
plt.tight_layout()

plt.show()
```

<h4>8.	Построить круговой график по данным п. 7</h4>

```python
plt.figure(figsize=(8, 8))
plt.pie(min_expenses, labels=years, autopct='%1.1f%%', startangle=140)
plt.title('Доля прожиточного минимума по годам (2012-2022)')
plt.axis('equal')


plt.show()
```

<h4>9.	Построить график произвольной каркасной поверхности.</h4>

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.cos(v), np.ones(np.size(u)))


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.plot_wireframe(x, y, z, color='b', linewidth=0.5)


ax.set_xlabel('X ось')
ax.set_ylabel('Y ось')
ax.set_zlabel('Z ось')
ax.set_title('Каркасная поверхность')


plt.show()
```
